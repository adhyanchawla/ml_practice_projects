```python
# Derivatives of x², x³, sin(x) — compute them numerically (f(x+h) - f(x)) / h and verify against analytical
# Partial derivatives for f(x,y) = x²y + 3y
# Chain rule: z = sin(x²), verify dz/dx numerically
# Build a numerical_gradient(f, x) utility function — you'll reuse this all week
```

```python
import numpy as np
# computing derivative as central difference
def numerical_derivative(f, x, h=1e-7):
    return (f(x + h) - f(x - h)) / (2 * h)
```

```python
f = lambda x: x**2
print(f"f(x) = x^2 at x = 3: numerical = {numerical_derivative(f, 3):.6f}, analytical = {2 * 3}")
```

```python
# forward difference is less accurate
def forward_derivative(f, x, h=1e-7):
    return (f(x + h) - f(x)) / h

f = lambda x: x**2
print(f"f(x) = x^2 at x = 3: numerical = {forward_derivative(f, 3):.6}")
```

```python
def numerical_gradient(f, params, h=1e-7):
    """Gradient of f wrt each element in params"""
    grad = np.zeros_like(params)
    for i in range(len(params)):
        old = params[i]

        params[i] = old + h
        loss_plus = f(params)

        params[i] = old - h
        loss_minus = f(params)

        grad[i] = (loss_plus - loss_minus) / (2 * h)
        params[i] = old

    return grad

def f(params):
    w, b = params[0], params[1]
    return w ** 2 + b ** 2

point = np.array([3.0, 2.0])
grad = numerical_gradient(f, point)

print(f"gradient: {grad}")
print(f"expected: [6.0, 4.0]")
```

```python
X = np.array([1,2,3,4,5],dtype=float)
y = np.array([6,11,16,21,26],dtype=float)

def mse_loss(params):
    w, b = params[0], params[1]
    predictions = w * X + b
    return np.mean((predictions - y) ** 2)

w, b = 0.0, 0.0
lr = 0.01

print(f"Start: w={w:.3f}, b={b:.3f}, loss={mse_loss(np.array([w,b])):.3f}")

for step in range(1000):
    params = np.array([w, b])
    grad = numerical_gradient(mse_loss, params)

    w = w - lr * grad[0] # nudge w downhill
    b = b - lr * grad[1] # nudge b downhill

    if step % 40 == 0:
        loss = mse_loss(np.array([w,b]))
        print(f"Step {step:3d}: w={w:.3f}, b={b:.3f}, loss={loss:.3f}")

print(f"\nFinal: w{w:.3f}, b={b:.3f}")
```
